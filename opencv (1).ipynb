```python
red = np.zeros((200,200,3));
red[:,:,0] = 0;
red[:,:,1] = 0;
red[:,:,2] = 200;
red = red.astype(np.uint8)
```

```python
green = np.zeros((200,200,3));
green[:,:,0] = 0;
green[:,:,1] = 100;
green[:,:,2] = 0;
green = green.astype(np.uint8)
```

```python
blue = np.zeros((200,200,3));
blue[:,:,0] = 200;
blue[:,:,1] = 0;
blue[:,:,2] = 0;
blue = blue.astype(np.uint8)
```

```python
merged = np.zeros((200,200,3));
merged[:,:,0] = 20;
merged[:,:,1] = 10;
merged[:,:,2] = 70;
merged = merged.astype(np.uint8)
```

```python
f = plt.figure(figsize=(15,15))
f.add_subplot(1,4,1).set_title('Red Image');
plt.imshow(red[:,:,::-1]);
f.add_subplot(1,4,2).set_title('Green Image');
plt.imshow(green[:,:,::-1]);
f.add_subplot(1,4,3).set_title('Blue Image');
plt.imshow(blue[:,:,::-1]);
f.add_subplot(1,4,4).set_title('Merged Image');
plt.imshow(merged[:,:,::-1]);
```

```python


sl1=widgets.IntSlider(value=100,min=1,max=255,step=1);
sl2=widgets.IntSlider(value=100,min=1,max=255,step=1);
sl3=widgets.IntSlider(value=100,min=1,max=255,step=1);

def f(red_c,green_c,blue_c):
  red=np.zeros((200,200,3));
  red[:,:,0]=0;
  red[:,:,1]=0;
  red[:,:,2]=red_c;
  red=red.astype(np.uint8)

  green=np.zeros((200,200,3));
  green[:,:,0]=0;
  green[:,:,1]=green_c;
  green[:,:,2]=0;
  green=green.astype(np.uint8)
  
  blue=np.zeros((200,200,3));
  blue[:,:,0]=blue_c;
  blue[:,:,1]=0;
  blue[:,:,2]=0;
  blue=blue.astype(np.uint8)

  merged=np.zeros((200,200,3));
  merged[:,:,0]=blue_c;
  merged[:,:,1]=green_c;
  merged[:,:,2]=red_c;
  merged=merged.astype(np.uint8)

  f=plt.figure(figsize=(15,15))
  f.add_subplot(1,4,1).set_title('Red Image');
  plt.imshow(red[:,:,::-1]);
  f.add_subplot(1,4,2).set_title('Green Image');
  plt.imshow(green[:,:,::-1]);
  f.add_subplot(1,4,3).set_title('Blue Image');
  plt.imshow(blue[:,:,::-1]);
  f.add_subplot(1,4,4).set_title('Merged Image');
  plt.imshow(merged[:,:,::-1]);
  plt.show()

interact(f,red_c=sl1,green_c=sl2,blue_c=sl3);
```
